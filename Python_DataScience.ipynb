### 3.2 Handlig Missing Data
Missing data conventions:
1) Usig a mask that globally idicates missing values.
2) Choosig a sentinel value that indicates a missing entry. Such as a number -999 or NaN.

The way in which Pandas handling missing values is constrained by its reliance on the Numpy package, which does not have a built-in notatio of NA values for non-floating-point data types. <br>

Numpy does have support for masked arrays. Arrays have a separate Boolean mask array attached for marking data "good" or "bad". Pandas could have derived from this, but the overhead in both storage, computation, and code maitenance makes that an unattractive choice. <br>

Pandas chose to use sentinels for missing data, either __NaN__ value (the special floating point) and the Python __None__ object. This choice has some side effects, but in practice ends up being a good compromise in most cases of interest.  <br>

__None__ is a Python object, it can't be used in any orbitrary Numpy/Pandas array, but only in arrays with data type object. While this kind of object may be useful for some purposes, any operation on the data will be done at the Python level, with much overhead than the typically fast operations seen for arrays with native types. If you perform aggregations like sum() or min() across array with a None value, you will generally get an error. <br>

__NaN__ is a special floating-point value recognized by all systems that use the standard IEEE floating-point representation. Regardless of the operation, the result of arithmetic with NaN will be another NaN. <br>

Numpy also provides some special aggregations that will ignore these missing values, for example, instead of using np.sum, use np.nansum. <br>

Pandas is built to handle None and NaN nearly interchangeably. <br>

There are several methods in detecting, removing, and replacing null values in Pandas data structures:
1) isnull()
generate a Boolean mask indicating missing values
2) notnull()
3) fillna()
Return a copy of the data with missing values filled or imputed.
4) dropna()

We can't drop single data in a DaraFrame. We can only drop full rows or full columns. <br>
dropna() will drop all rows with any null value is present.. <br>
dropna(axis = 1) / dropna(axis='column') drops all columns containing a null value. <br>
dropna(how='all') will drop rows with all null values. <br>
dropna(thresh=3) the thresh parameters let you specify a minimum number of non-null values for the row/column to be kept.<br>

Sometimes rather than dropping NA values, you'd rather replace them with a valid value, such as 0.

### Visualization with Seaborn
Matplotlib has proven to be useful, but there are several valid comlaints about it: 
1) Matplotlib's API is relatively low level. Doing sophiscated statistical visulization is possible, but often requires a lot of boilerplate code.
2) Matplotlib predated Pandas by more than a decade, and thus is not designed for use with Pandas Dataframes. In order to visulize data from a Pandas DataFrame, you must extract each Series and concatenate them together into the right format. It will be nicer to have a plotting library that can intelligently use the DataFrame labels in a plot.

An answer to these probelms is Seaborn. Seaborn provides an API on top of Matplotlib that offers same choices for plot style and color defaults, defines simple high-level functions for common statistical plot styles, and integrates with the functionality provided by Pandas DataFrames. <br>

Matplotlib team has recently added the plt.style tools, and it is starting to handle Pandas data more seamlessly. However, Seaborn remains an extremely useful add-on. <br>

The steps of using a Scikit-Learn API are as follows:
- Choose a class of model and model parameters. Note a class of model is not the same as an instance of a model.

- Fit the model to your data by calling the fit()
- To apply the model to new data
    - For supervised learning, use predict()
    - For unsupervised learning, use transform() or predict()

```python
from sklearn.decomposition import PCA
model = PCA(n_components = 2)
model.fit(x_iris)
x_iris_2D = model.transform(x_iris) # transform the data to 2D
```

__Unsupervised learning: iris clustering__

GMM (Gaussian mixture model): a powerful clustering method. A GMM attempts to model the data as a collection of Gaussian blobs. <br>

```python
from sklearn import mixture # choose the model class
model = mixture.GaussianMixture(n_components = 3, covariance_type = 'full') # instantiate the model with hyperparameters
model.fit(x_iris) # feature matrix, without labels
y_gmm = model.predict(x_iris) # determine cluster labels
```

```python
# We will then add the cluster labels to the Iris DataFrame and use SeaBorn to plot the results
iris['cluster'] = y_gmm
sns.lmplot(data=iris, x='PCA1',y='PCA2', hue='species', col = 'cluster')
```

By splitting the data by cluster number, we see exactly how well the GMM algorithm has recovered the underlying label: the setosa species is separated perfectly within cluster 0, whiler a small amount of mixung between versicolor and virginica. This means that even without an expert to tell us the species labels of the individual flowers, the measurements of these flowers are distinct enough that we could automatically identify the presence of these different groups of species with a simple clustering algorithm.

### Naive Bayes Classification

Naive Bayes models are a group of extremely fast and simple classification algorithms that are often suitable for very high-dimensional datasets. It is built upon Baysian classification models. They rely on Bayes's theorem:

$$ P(L|features) = \frac{P(features|L) \cdot P(L)}{P(features)} $$

If we are trying to decide between two labels, L1 and L2, then one way to make decision is to compute the ratio of the posterior probabilities for each label:
$$ \frac{P(L_1|features)}{P(L_2|features)} = \frac{P(features|L_1) \cdot P(L_1)}{P(features|L_2) \cdot P(L_2)} $$

All we need now is some model that we can compute $P(features|L_1)$ and $P(features|L_2)$. Such a model is called a generative model. <br>

Naive Bayes: we make naive assumptions about the models for each label. Different types of naive Bayes classifiers rest on different naive assumptions about the data. Gaussian Naive Bayes has the assumption that data from each label is drawn from a simple Gaussian distribution with no covariance between dimensions. <br>

Multinomial Naive Bayes: the features are assumped to be generated from a simple multinomial distribution. 

### Linear Regression

Just as naive Bayes is a good starting point for classification tasks, linear regression models are a good starting point for regression tasks. Such models are popular because they can be fit very quickly, and are very interpretable. <br>

Linear regression can be one dimensional or multidimensional. <br>
from sklearn.linear_model import LinearRegression <br>

__`Basis Function Regression`__ <br>
One way you can use to adapt linear regression to nolinear relationship between variables is to transform the data according to basis functions. <br>

$ y = a_0 + a_1x_1 + a_2x_2 + a_3x_3 + ...$ and $ y = a_0 + a_1x + a_2x^2 + a_3x^3 + ...$ are the same if you consider $x_2 = x^2$, $x_3 = x^3$ <br>

Polynomial basis function use a sum of polynomial bases. It is built into SciKit-Learn. PolynomialRegression pipeline <br>
Gaussian basis function: another useful pattern is to fit a model using a sum of Gaussian bases. The Gaussian basis functionsare not bbuilt into Scikit-Learn, but we can write a custom transformer that will create them. Scikit-Learn transformers are implemented as Python classes. <br>

Classes provide a means of bundling data and functionality together. Classes can have attributes and methods. <br>

__`Regularization`__ <br>
The introduction of basis functions into our linear regression makes the model much more flexible, but it also can very quickly lead to overfitting. For example, if we choose too many Gaussian basis functions, we end up with results that don't look so good.

We don't want spikes (large values) of the model parameters. A penalty in the model tp constrain the large values of the model parameters is known as regularization, and comes in several forms. <br>

__Ridge regularization ($L_2$ regularization)__ (Tikhonov regularization) <br>
It penalize the sum of squares (2-norm) of the model coefficients. The penalty on the model fit would be:
$$ P = \alpha \sum_{n=1}^N \theta_n^2$$
where $\alpha$ is a free parameter that controls the strength of the penalty. $\alpha$ close to 0, we recover the standard linear regression. When $\alpha$ close to infinity, all model responses will be suppressed. $L_2$ regularization is built into SciKit-Learn with the Ridge estimator. One advantage of $L_2$ regularization is that it can be computed very efficiently. <br>

__Lasso regularization ($L_1$ regularization)__ <br>
Lasso regularization penalizes the sum of absolute values (1-norm) of regression coefficients:
$$ P = \alpha \sum_{n=1}^N |\theta_n|$$
Due to geometric reasons, Lasso regresssion tends to favor sparse models where possible; that is, it preferetially sets model coefficients to exactly zero.

### Support Vector Machines (SVMs)
Support vector machines are particularly powerful and flexible class of supervised algorithms for both classification and regression. <br>

Earlier in the Naive Bayes Classification, we model each class. Hererather than modelling each class, we simply find aline or curve (in 2D space) or manifold (in multiple dimensions) that divideds the classes from each other. However, there seems many ways of drawing a line. SVM offers a solution: instead of drwing a single line, we can draw a line with some width (like an error of the intercept). The line that maximize this margine is the one we will choose as the optimal model. <br>

SVM is extremely powerful when it is combined with kernels. <br>
SVC(kernel='linear') <br>
SVC(kernel='rbf') (radial basis function) <br>

If your data is not perfectly clean, and shows some kind of overlap, a perfect decision boundary may not exit. In this case, the SVM implementation has a bit of fudge-factor that softens the margin. It allows some of the points to creep into the margin if that allows a better fit. The hardness of a margin is controlled by a tuning parameter, C. For a large C, the margin is hard, and points can't lie in it. The optimal value of C will depend on your dataset, and should be tuned via cross-validation or a similar procedure. <br>

```python
def plot_svc_decision_function(model, ax=None, plot_support=True):
    """Plot the decision function for a two-dimensional SVC"""
    if ax is None:
        ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    # create grid to evaluate model
    x = np.linspace(xlim[0], xlim[1], 30)
    y = np.linspace(ylim[0], ylim[1], 30)
    Y, X = np.meshgrid(y, x)
    xy = np.vstack([X.ravel(), Y.ravel()]).T
    P = model.decision_function(xy).reshape(X.shape)
    
    # plot decision boundary and margins
    
    ax.contour(X, Y, P, colors='k',
    levels=[-1, 0, 1], alpha=0.5,
    linestyles=['--', '-', '--'])
    
    # plot support vectors
    if plot_support:
        ax.scatter(model.support_vectors_[:, 0],
            model.support_vectors_[:, 1],
            s=300, linewidth=1, facecolors='none', edgecolors='k');
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
```

### k-means Clustering
Clustering algorithms seek to learn, from the properties of the data, an optimal division or discrete labeling of groups of points. Many clustering algorithms are available in Scikit-Learn and elsewhere, but perhaps the simplest to understand is an algorithm knowns as k-means clustering, which is implemented in sklearn.cluster.KMeans. <br>

The cluster center is the arithmetic mean of all the points belonging to the cluster. Each point is closer to its own cluster center than to other cluster centers. These two assumptions are the basis of the k-means model. <br>

__`k-means algorithm: Expectation-Maximization`__
- Guess some cluster centers
- Repeat until converged
    - E-step: assign points to the nearest cluster center
    - M-step: set the cluster centers to the mean

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=200, centers=4, cluster_std=0.5, random_state=0)
plt.scatter(X[:,0], X[:,1], c=y, cmap='winter')
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4,n_init='auto')
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

```python
plt.scatter(X[:,0], X[:,1], c=y_kmeans, cmap='winter')
centers = kmeans.cluster_centers_
plt.scatter(centers[:,0], centers[:,1], c='red', s=200)
```
